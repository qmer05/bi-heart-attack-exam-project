```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = '../data/cleaned_digital_diet_mental_health.csv'
df = pd.read_csv(dataset, header=0)
```

```python
# Fjern binære one-hot kolonner med kun to unikke værdier
numeric_cols = df.select_dtypes(include='number').columns
filtered_cols = [col for col in numeric_cols if df[col].nunique() > 2]

# Histogram kun på relevante kolonner
df[filtered_cols].hist(bins=20, figsize=(15, 10), grid=False)
plt.tight_layout()
plt.show()
```

```python
# Genskab kategorien "gender" fra one-hot
gender_cols = [col for col in df.columns if col.startswith('gender_')]
df['gender'] = df[gender_cols].idxmax(axis=1).str.replace('gender_', '')

# Samlet visualisering
df['gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
location_cols = [col for col in df.columns if col.startswith('location_type_')]
df['location_type'] = df[location_cols].idxmax(axis=1).str.replace('location_type_', '')

df['location_type'].value_counts().plot(kind='bar')
plt.title('Location Type Distribution')
plt.xlabel('Location Type')
plt.ylabel('Count')
plt.show()
```

```python
# Fjern binære one-hot kolonner med kun to unikke værdier
numeric_cols = df.select_dtypes(include='number').columns
filtered_cols = [col for col in numeric_cols if df[col].nunique() > 2]

for col in filtered_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
```

```python
# Genskab 'gender' kolonne fra one-hot
gender_cols = [col for col in df.columns if col.startswith('gender_')]
df['gender'] = df[gender_cols].idxmax(axis=1).str.replace('gender_', '')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.histplot(
    data=df,
    x='stress_level',
    hue='gender',
    multiple='dodge',
    shrink=0.8,
    discrete=True  # stress_level er diskret
)
plt.title('Stress Level Distribution by Gender')
plt.xlabel('Stress Level')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_stress = df.groupby('gender')['stress_level'].mean()
print("Gennemsnitligt stressniveau per køn:")
print(avg_stress)

diff = avg_stress.max() - avg_stress.min()
print(f"\nForskellen mellem højeste og laveste gennemsnit: {diff:.2f}")
```

```python
# Genskab 'location_type' fra one-hot kolonner
location_cols = [col for col in df.columns if col.startswith('location_type_')]
df['location_type'] = df[location_cols].idxmax(axis=1).str.replace('location_type_', '')

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.histplot(
    data=df,
    x='stress_level',
    hue='location_type',
    multiple='dodge',
    shrink=0.8,
    discrete=True  # stress_level er heltal
)
plt.title('Stress Level Distribution by Location Type')
plt.xlabel('Stress Level')
plt.ylabel('Frequency')
plt.show()

```

```python
avg_stress_loc = df.groupby('location_type')['stress_level'].mean()
print("Gennemsnitligt stressniveau pr. bopælstype:")
print(avg_stress_loc)

diff = avg_stress_loc.max() - avg_stress_loc.min()
print(f"\nForskel mellem højeste og laveste gennemsnit: {diff:.2f}")
```

```python
df['depression_level'] = pd.cut(
    df['weekly_depression_score'],
    bins=[-1, 5, 15, 30],  # justér efter dit datasæt
    labels=['Low', 'Moderate', 'High']
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.histplot(
    data=df,
    x='daily_screen_time_hours',
    hue='depression_level',  # Bruger kategorierne
    multiple='stack',
    bins=20
)
plt.title('Daily Screen Time by Depression Level')
plt.xlabel('Screen Time (hours/day)')
plt.ylabel('Frequency')
plt.show()
```
