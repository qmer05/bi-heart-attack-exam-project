### Standard scaling

```python
# reduce all with the mean and scale the data to unit variance
# x = (x-xmean)/std
standard_scaler = StandardScaler()
scaled_data['age'] = standard_scaler.fit_transform(scaled_data[['age']])

print('Mean:', scaled_data['age'].mean()) # almost 0
print('Standard Deviation:', scaled_data['age'].std()) # almost 1
```

```python
# histogram has same shape, but 0,0 is in the middle
sb.histplot(scaled_data['age'], color='#ee4c2c', bins=50);
```

```python
minmax_scaler = MinMaxScaler()
scaled_data['body_mass_min_max_scaled'] = minmax_scaler.fit_transform(scaled_data[['age']])

print('Mean:', scaled_data['age'].mean())
print('Standard Deviation:', scaled_data['body_mass_min_max_scaled'].std())
```
